```python
import pandas as pd 
import numpy as np
import re
from helper import *
import sqlite3
```

## LinkedIn Data

```python
linkedin_data = pd.read_csv('data\ML engineer jobs (aug 30, 2025) - Linkedin.csv')

linkedin_data = clean_job_data(linkedin_data, "LinkedIn")

linkedin_data.head(2)
```

## Glints Data

```python
glints_data = pd.read_csv("data\ML engineer jobs (aug 30, 2025) - Glints.csv")

glints_data = clean_job_data(glints_data, "Glints")

glints_data.head(2)
```

## Dealls Data

```python
dealls_data = pd.read_csv("data\ML engineer jobs (aug 30, 2025) - dealls.csv")
dealls_data = clean_job_data(dealls_data, "Dealls")

dealls_data.head(2)
```

## Kalibrr Data

```python
kalibrr_data = pd.read_csv("data\ML engineer jobs (aug 30, 2025) - Kalibrr.csv")
kalibrr_data = clean_job_data(kalibrr_data, "Kalibrr")

kalibrr_data.head()
```

```python
indeed_data = pd.read_csv("data\ML engineer jobs (aug 30, 2025) - Indeed.csv")
indeed_data = clean_job_data(indeed_data, "Indeed")

indeed_data
```

```python
dfs = [linkedin_data, glints_data, dealls_data, kalibrr_data, indeed_data]

all_jobs = pd.concat(dfs, ignore_index= True)

```

## Create SQLite Database

```python
conn = sqlite3.connect("jobs.db")

all_jobs = all_jobs.drop(columns="Salary", axis = 1)

all_jobs.to_sql("jobs", conn, if_exists="replace", index = False)
```
